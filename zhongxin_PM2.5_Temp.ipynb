```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
%matplotlib inline

dateparse = lambda x: pd.datetime.strptime(x, '%d/%m/%Y %H:%M') 
# Read data from file
data = pd.read_csv("zhongxin_data_2013_15.csv", usecols=['date', 'PM2_5', 'NO2', 'CO', 'TEMP'],
                   parse_dates=['date'], date_parser=dateparse)
data['date']=pd.to_datetime(data.date)

# Preview the data 
data.head()
```

```python
# drop all rows with NaN
data = data.dropna()
```

```python
X = np.array(data['TEMP'].dropna().values, dtype=float).reshape(-1, 1)
y = np.array(data['PM2_5'].dropna().values, dtype=float).reshape(-1, 1)

plt.figure(figsize=(15,10))
plt.scatter(X, y, color='red')
plt.title('Temperature and PM2.5')
plt.xlabel('Temperature')
plt.ylabel('PM2.5')
plt.show()
```
